```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tb = pd.read_csv('BaseFeminicidioEvolucaoMensalCisp.csv', sep = ';', encoding="UTF-8")

#REMOVE AS COLUNAS
tb = tb.drop(columns = ['CISP', 'AISP', 'RISP', 'Fase'] )

#2023 à 2026

#FILTRA AS COLUNAS
#SALVO EM VARIÁVEIS PRA NAO BAGUNÇAR E PRA APLICAR FILTROS MAIS ESPECIFICOS
#POR EXEMPLO NÃO DA PRA FILTRAR OS DOIS EXEMPLOS JUNTOS
#POIS UM TEM QUE SER MAIOR QUE 0 E O OUTRO 2023
filtroAno2023 = (tb['Ano'] > 2022)
filtroFemi2023 = (tb['Feminic�dio']) | (tb['Tentativa de feminic�dio'] > 0)
#VARIAVEL CRIADA PARA SALVAR A TABELA FILTRADA
tbFiltrado2023 = tb[filtroAno2023 & filtroFemi2023]


#pd.to_numeric PEGA A COLUNA 'ANO' E TRANSFORMA EM INTEIRO
#np.array TRANSFORMA ESSES NUMEROS NUM ARRAY E REMOVE OS NUMEROS DUPLICADOS
Anos2023 = pd.to_numeric(tb['Ano'], errors='coerce')
Anos2023 = np.array(Anos2023.drop_duplicates())
Anos2023 = Anos2023[Anos2023 > 2022]

#Total de feminicidios e tentativas por ano 2023 a 2026
#np.flip para inverter as arrays para acompanhar a forma em que os anos são salvos
#Também para o gráfico ficar certo
FeminPAno2023 = np.array([99, 107, 105, 12])
FeminPAno2023 = np.flip(FeminPAno2023)

TentaPAno2023 = np.array([308, 382, 307, 57])
TentaPAno2023 = np.flip(TentaPAno2023)

#AQUI SOMA OS FEMINICIDIOS E AS TENTATIVAS JÁ FILTRADAS DA VARIAVEL tbFiltrado2023
totalFemin2023 = tbFiltrado2023['Feminic�dio'].sum()
totalTenta2023 = tbFiltrado2023['Tentativa de feminic�dio'].sum()


#AQUI PRINTA SO KK
#print(tbFiltrado)
#print("Feminicidios no período de 2023 à 2026 no estado do Rio de Janeiro:", totalFemin2023)
#print("Tentativas de Feminicidios no período de 2023 à 2026 no estado do Rio de Janeiro:", totalTenta2023)


#USANDO O NUMPY PARA CALCULAR O INTERCEPTO E COEFICIENTE ANGULAR
cAngularFemin2023, interFemin2023 = np.polyfit(Anos2023, FeminPAno2023, 1)

#CALCULANDO O PREVISTO
FeminPred2023 = cAngularFemin2023 * Anos2023 + interFemin2023

#FAZENDO O GRÁFICO DA REGRESSÃO LINEAR
plt.scatter(Anos2023, FeminPAno2023, label = 'Dados Originais')
plt.plot(Anos2023, FeminPred2023, color = 'red', linewidth = 2, label = 'Linha de Regressão')

#ESSA FUNÇÃO REMOVE OS NÚMEROS QUEBRADOS NO EIXO X, POR EXEMPLO, ANTES HAVIA 2025.5
plt.xticks(np.arange(min(Anos2023), max(Anos2023)+1, 1.0))

plt.xlabel('Anos 2023-2026')
plt.ylabel('Total de Femincidios 2023-2026')
plt.legend()
plt.title('Regressão Linear do total de Femincídios entre 2023 à 2026')
plt.grid(True)


for i in range(Anos2023.size):
  print('Feminicidios nos anos de',Anos2023[i], ':', FeminPAno2023[i])
```

```python
#REGRESSÃO LINEAR TENTATIVA


cAngularTenta2023, interTenta2023 = np.polyfit(Anos2023, TentaPAno2023, 1)

TentaPred2023 = cAngularTenta2023 * Anos2023 + interTenta2023

plt.scatter(Anos2023, TentaPAno2023, label = 'Dados Originais')
plt.plot(Anos2023, TentaPred2023, color = 'red', linewidth = 2, label = 'Linha de Regressão')

#ESSA FUNÇÃO REMOVE OS NÚMEROS QUEBRADOS NO EIXO X, POR EXEMPLO, ANTES HAVIA 2025.5
plt.xticks(np.arange(min(Anos2023), max(Anos2023)+1, 1.0))

plt.xlabel('Anos 2023-2026')
plt.ylabel('Total de Tentativas de Femincidios 2023-2026')
plt.legend()
plt.title('Regressão Linear do total de Tentativas de Femincídios entre 2023 à 2026')
plt.grid(True)

for i in range(Anos2023.size):
  print('Tentativas no ano de',Anos2023[i], ':', TentaPAno2023[i])
```
